```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("employee_attendance.csv")
df.head()
```

```python

df['Date'] = pd.to_datetime(df['Date'])
df['InTime'] = pd.to_datetime(df['InTime'], format="%H:%M")
df['OutTime'] = pd.to_datetime(df['OutTime'], format="%H:%M")


df.head()
```

```python
df['WorkingHours'] = (df['OutTime'] - df['InTime']).dt.seconds / 3600
df[['Name', 'Date', 'InTime', 'OutTime', 'WorkingHours']]
```

```python
print("----- Summary of Working Hours -----")
print(df['WorkingHours'].describe())
```

```python
avg_hours = df.groupby("Name")['WorkingHours'].mean()
print(avg_hours)
```

```python
plt.figure(figsize=(8,5))
avg_hours.plot(kind='bar')
plt.xlabel("Employee")
plt.ylabel("Average Working Hours")
plt.title("Average Working Hours Per Employee")
plt.show()
```

```python
df['IsLate'] = df['InTime'].dt.time > pd.to_datetime("09:30", format="%H:%M").time()
df[['Name', 'Date', 'InTime', 'IsLate']]
```

```python
late_count = df.groupby("Name")['IsLate'].sum()
print(late_count)
```
